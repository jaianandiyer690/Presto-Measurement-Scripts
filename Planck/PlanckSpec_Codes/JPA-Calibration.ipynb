```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
import ipywidgets as widgets

plt.rcParams['axes.formatter.useoffset'] = False
plt.rcParams['figure.constrained_layout.use'] = True
plt.rcParams['axes.labelsize'] = 17
plt.rcParams['font.size'] = 15
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['legend.frameon'] = False
```

```python
def dB(Sjk):
    return 20 * np.log10(np.abs(Sjk))
```

### 2022-11-18_13_59_31

```python
# Load data
file = r'D:\JPA-Data\2022-11\2022-11-test.hdf5'
meas_type = 'calibration'
run = '2022-11-18_16_49_08'
idx_str = "{}/{}".format(meas_type, run)

# Open hdf5 file
with h5py.File(file, "r") as dataset:
    # Data
    freq_arr = np.asarray(dataset[idx_str]["freq sweep"])
    usb_arr = np.asarray(dataset[idx_str]["USB"])
    

fig, ax = plt.subplots(1)
ax.plot(freq_arr/1e9, dB(usb_arr), '.-')
ax.set_xlabel('frequency [GHz]')
ax.set_ylabel('signal magnitude [dBFS]')
plt.show()
```

```python
# Load data
file = r'D:\JPA-Data\2022-11\2022-11-test.hdf5'
meas_type = 'calibration'
run = '2022-11-18_16_29_08'
idx_str = "{}/{}".format(meas_type, run)

# Open hdf5 file
with h5py.File(file, "r") as dataset:
    # Data
    freq_arr = np.asarray(dataset[idx_str]["freq sweep"])
    usb_arr_ref = np.asarray(dataset[idx_str]["USB"])
    

fig, ax = plt.subplots(1)
ax.plot(freq_arr/1e9, dB(usb_arr_ref), '.-')
ax.set_xlabel('frequency [GHz]')
ax.set_ylabel('signal magnitude [dBFS]')
plt.show()
```
